```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('stock.csv', sep=';')
```

```python
plt.figure(figsize=(20, 6))
plt.hist(df['Purchase Price'], bins=24, edgecolor='green')
plt.xlabel('Purchase Price')
plt.ylabel('Count')
plt.title('Grafik Histogram')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Current Price'], bins=24, edgecolor='green')
plt.xlabel('Current Price')
plt.ylabel('Count')
plt.title('Grafik Histogram')
plt.show()
```
